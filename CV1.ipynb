```python
# --- LIBRARY INSTALLS --- ### Mohsen Mohammadagha
import sys, subprocess
def install_if_missing(pkg, import_name=None):
    import_name = import_name or pkg
    try:
        __import__(import_name)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])

for pkg, alias in [("tensorflow", None), ("numpy", None), ("pandas", None), ("matplotlib", None),
                   ("seaborn", None), ("scikit-learn", "sklearn"), ("opencv-python", "cv2"),
                   ("xgboost", None)]:
    install_if_missing(pkg, alias)

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns, cv2, tensorflow as tf
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import (classification_report, confusion_matrix, precision_score,
                             recall_score, f1_score, roc_curve, auc, roc_auc_score)
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import label_binarize
from sklearn.manifold import TSNE
import xgboost as xgb
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Reshape, BatchNormalization,
                                     Lambda, GlobalAveragePooling2D)
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img

# ================
# RANKED HARDWARE SELECTION AND STRATEGY
# ================
hardware_list = [
    "v6e-1 TPU",
    "A100 GPU",
    "L4 GPU",
    "v5e-1 TPU",
    "v2-8 TPU",
    "T4 GPU",
    "CPU"
]

def is_available(hardware):
    if "TPU" in hardware:
        try:
            tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
            return tpu is not None
        except Exception:
            return False
    elif "A100" in hardware or "L4" in hardware or "T4" in hardware or "GPU" in hardware:
        gpus = tf.config.list_physical_devices('GPU')
        return len(gpus) > 0
    elif "CPU" in hardware:
        return True
    return False

selected_hardware = None
for hw in hardware_list:
    if is_available(hw):
        selected_hardware = hw
        break

if selected_hardware:
    print(f"Selected hardware: {selected_hardware}")
else:
    print("No hardware available from the list.")

try:
    strategy = None
    if "TPU" in selected_hardware:
        tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
        tf.config.experimental_connect_to_cluster(tpu)
        tf.tpu.experimental.initialize_tpu_system(tpu)
        strategy = tf.distribute.TPUStrategy(tpu)
        print("Using TPU strategy")
    elif "GPU" in selected_hardware and tf.config.list_physical_devices('GPU'):
        strategy = tf.distribute.MirroredStrategy()
        print("Using GPU strategy")
    else:
        strategy = tf.distribute.get_strategy()
        print("Using CPU/default strategy")
except Exception as e:
    print("Accelerator detection error:", e)
    strategy = tf.distribute.get_strategy()

# ================
# 1. DATA LOAD & CLASSIC ML BASELINES
# ================
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
X = lfw_people.data
y = lfw_people.target
target_names = lfw_people.target_names
h, w = lfw_people.images.shape[1:3]
print(f"Dataset loaded: {X.shape[0]} samples, {X.shape[1]} features")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
dims = range(20, 141, 20)
accs_svm = []; accs_ann = []; accs_xgb = []
roc_svm = []; roc_ann = []; roc_xgb = []

y_test_bin_full = label_binarize(y_test, classes=np.arange(len(target_names)))

for n in dims:
    pca = PCA(n_components=n, whiten=True, random_state=42)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    # SVM
    clf = SVC(kernel='rbf', probability=True, random_state=42) #C=0.1) ########
    clf.fit(X_train_pca, y_train)
    accs_svm.append(clf.score(X_test_pca, y_test)*100)
    svm_proba = clf.predict_proba(X_test_pca)
    roc_auc = roc_auc_score(y_test_bin_full, svm_proba, average='macro', multi_class='ovr')
    roc_svm.append(roc_auc)
    # ANN
    ann = Sequential([
        Dense(256, activation='relu', input_shape=(n,)), Dropout(0.3),
        Dense(128, activation='relu'), Dense(len(target_names), activation='softmax')
    ])
    ann.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    ann.fit(X_train_pca, to_categorical(y_train), epochs=10, batch_size=64, verbose=0)
    eval_ann = ann.evaluate(X_test_pca, to_categorical(y_test), verbose=0)
    accs_ann.append(eval_ann[1]*100)
    ann_proba = ann.predict(X_test_pca)
    roc_auc_ann = roc_auc_score(y_test_bin_full, ann_proba, average='macro', multi_class='ovr')
    roc_ann.append(roc_auc_ann)
    # XGBoost
    xgb_model = xgb.XGBClassifier(eval_metric='mlogloss', use_label_encoder=False, verbosity=0)
    xgb_model.fit(X_train_pca, y_train)
    accs_xgb.append(xgb_model.score(X_test_pca, y_test)*100)
    xgb_proba = xgb_model.predict_proba(X_test_pca)
    roc_auc_xgb = roc_auc_score(y_test_bin_full, xgb_proba, average='macro', multi_class='ovr')
    roc_xgb.append(roc_auc_xgb)

plt.figure(figsize=(8,6))
plt.plot(dims, accs_svm, '-o', label='SVM+PCA (Acc)')
plt.plot(dims, accs_ann, '-o', label='ANN+PCA (Acc)')
plt.plot(dims, accs_xgb,'-o',label='XGBoost+PCA (Acc)')
plt.xlabel('Reduced Number of Dimension'); plt.ylabel('Recognition Rate (%)')
plt.title('Recognition Rate vs. Dimensionality')
plt.legend(); plt.tight_layout(); plt.show()

plt.figure(figsize=(8,6))
plt.plot(dims, roc_svm, '-o', label='SVM+PCA (ROC-AUC)')
plt.plot(dims, roc_ann, '-o', label='ANN+PCA (ROC-AUC)')
plt.plot(dims, roc_xgb,'-o',label='XGBoost+PCA (ROC-AUC)')
plt.xlabel('Reduced Number of Dimension'); plt.ylabel('Macro ROC AUC')
plt.title('Macro ROC AUC vs. Dimensionality')
plt.legend(); plt.tight_layout(); plt.show()

# Meta model: stacking (SVM, XGBoost)
pca = PCA(n_components=dims[-1], whiten=True, random_state=42)
X_train_pca = pca.fit_transform(X_train); X_test_pca = pca.transform(X_test)
clf = SVC(kernel='rbf', probability=True, random_state=42) #C=0.1) ###########
clf.fit(X_train_pca, y_train)
xgb_model = xgb.XGBClassifier(eval_metric='mlogloss', use_label_encoder=False, verbosity=0)
xgb_model.fit(X_train_pca, y_train)
stacking = StackingClassifier([('svm', clf), ('xgb', xgb_model)],
                              final_estimator=LogisticRegression(), passthrough=False, cv=5)
stacking.fit(X_train_pca, y_train)
stack_acc = stacking.score(X_test_pca, y_test)*100
stack_proba = stacking.predict_proba(X_test_pca)
roc_auc_stack = roc_auc_score(y_test_bin_full, stack_proba, average='macro', multi_class='ovr')
print(f'Stacking meta-model test accuracy (last dim): {stack_acc:.2f}%')
print(f'Stacking meta-model macro ROC-AUC (last dim): {roc_auc_stack:.3f}')

# ROC CURVES for all models (last dimension)
plt.figure(figsize=(8,6))

clf_svm_plot = SVC(kernel='rbf', probability=True, random_state=42)
clf_svm_plot.fit(X_train_pca, y_train)
proba_svm = clf_svm_plot.predict_proba(X_test_pca)
fpr, tpr, _ = roc_curve(y_test_bin_full.ravel(), proba_svm.ravel())
plt.plot(fpr, tpr, label=f'SVM+PCA (AUC={roc_auc_score(y_test_bin_full, proba_svm, average="macro", multi_class="ovr"):.2f})')

ann_plot = Sequential([
    Dense(256, activation='relu', input_shape=(dims[-1],)), Dropout(0.3),
    Dense(128, activation='relu'), Dense(len(target_names), activation='softmax')
])
ann_plot.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
ann_plot.fit(X_train_pca, to_categorical(y_train), epochs=10, batch_size=64, verbose=0)
proba_ann = ann_plot.predict(X_test_pca)
fpr, tpr, _ = roc_curve(y_test_bin_full.ravel(), proba_ann.ravel())
plt.plot(fpr, tpr, label=f'ANN+PCA (AUC={roc_auc_score(y_test_bin_full, proba_ann, average="macro", multi_class="ovr"):.2f})')

xgb_plot = xgb.XGBClassifier(eval_metric='mlogloss', use_label_encoder=False, verbosity=0)
xgb_plot.fit(X_train_pca, y_train)
proba_xgb = xgb_plot.predict_proba(X_test_pca)
fpr, tpr, _ = roc_curve(y_test_bin_full.ravel(), proba_xgb.ravel())
plt.plot(fpr, tpr, label=f'XGBoost+PCA (AUC={roc_auc_score(y_test_bin_full, proba_xgb, average="macro", multi_class="ovr"):.2f})')

stacking_plot = StackingClassifier([('svm', clf_svm_plot), ('xgb', xgb_plot)],
                                   final_estimator=LogisticRegression(), passthrough=False, cv=5)
stacking_plot.fit(X_train_pca, y_train)
proba_stack = stacking_plot.predict_proba(X_test_pca)
fpr, tpr, _ = roc_curve(y_test_bin_full.ravel(), proba_stack.ravel())
plt.plot(fpr, tpr, label=f'Stacking (AUC={roc_auc_score(y_test_bin_full, proba_stack, average="macro", multi_class="ovr"):.2f})')

plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title('ROC Curves (All Classic Models, OvR, Last Dim)')
plt.legend(); plt.show()

# ================
# 2. SVM+PCA DETAILED EVALUATION AND VISUALIZATIONS
# ================

n_components = 150
pca = PCA(n_components=n_components, whiten=True, random_state=42)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

clf = SVC(kernel='rbf', class_weight='balanced', probability=True, random_state=42)
clf.fit(X_train_pca, y_train)
y_pred = clf.predict(X_test_pca)
print("=== SVM Classification Report ===")
print(classification_report(y_test, y_pred, target_names=target_names))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Confusion Matrix Heatmap

mask = np.triu(np.ones_like(cm, dtype=bool))  # For upper triangle, use np.tril

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='inferno', mask=mask)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('SVM Confusion Matrix Heatmap')
plt.show()

# Normalized Confusion Matrix
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(8,6))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='RdYlBu_r')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('SVM Normalized Confusion Matrix')
plt.show()

# Per-Class Precision, Recall, F1 Score Bar Chart
precision = precision_score(y_test, y_pred, average=None)
recall = recall_score(y_test, y_pred, average=None)
f1 = f1_score(y_test, y_pred, average=None)
x = range(len(target_names))
plt.bar(x, precision, width=0.2, label='Precision')
plt.bar([i+0.2 for i in x], recall, width=0.2, label='Recall')
plt.bar([i+0.4 for i in x], f1, width=0.2, label='F1 Score')
plt.xticks([i+0.2 for i in x], target_names, rotation=45)
plt.legend()
plt.title('SVM Per-Class Precision, Recall, F1 Score')
plt.tight_layout()
plt.show()

# Support (Sample Count) Bar Chart
unique, counts = np.unique(y_test, return_counts=True)

cmap = plt.get_cmap('viridis')
inferno_colors = cmap(np.linspace(0, 1, len(target_names)))

colors = cmap(np.linspace(0, 1, len(target_names)))  # Creates a gradient of colors

plt.figure(figsize=(8,6))
plt.bar(target_names, counts, width=0.5, color=colors)
plt.xticks(rotation=45)
plt.ylabel('Sample Count', fontsize=12, color='green')
plt.title('SVM Number of Samples per Class', color='blue', fontsize=16) #######
#plt.xticks(rotation=45, fontsize=12, color='red')  # X-tick labels

ax = plt.gca()
for label, color in zip(ax.get_xticklabels(), inferno_colors):
    label.set_color(color)
    label.set_fontsize(12)
plt.xticks(rotation=45)



plt.yticks(fontsize=12, color='purple')              # Y-tick labels
plt.tight_layout()
plt.show()




# Error Analysis Table
cm_df = pd.DataFrame(cm, index=target_names, columns=target_names)
errors = []
for i, true_label in enumerate(target_names):
    row = cm_df.iloc[i].copy()
    row[i] = 0  # Ignore correct predictions
    if row.sum() > 0:
        most_common_wrong = row.idxmax()
        errors.append((true_label, most_common_wrong, row.max()))
error_df = pd.DataFrame(errors, columns=['True Label', 'Most Common Mistake', 'Count'])
print("Most Common Misclassifications per Class (SVM):")
print(error_df)

# ROC Curve (OvR)
y_test_bin = label_binarize(y_test, classes=range(len(target_names)))
y_score = clf.decision_function(X_test_pca)
plt.figure(figsize=(8,6))
for i in range(len(target_names)):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    plt.plot(fpr, tpr, label=f'ROC curve of class {target_names[i]}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('SVM ROC Curve (One-vs-Rest)')
plt.legend(loc='lower right')
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score, average='macro', multi_class='ovr')
print(f"SVM AUC Score: {auc_score:.2f}")

# t-SNE Embedding Visualization (subset for speed)
subset_size = 300
X_test_subset = X_test_pca[:subset_size]
y_test_subset = y_test[:subset_size]
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(X_test_subset)
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y_test_subset, cmap='Pastel1') #tab10
plt.colorbar(scatter, ticks=range(len(target_names)), label='Class')
plt.title('SVM t-SNE Projection of Test Set')
plt.show()

# Learning Curve
train_sizes, train_scores, test_scores = learning_curve(clf, X_train_pca, y_train, cv=5)
plt.figure(figsize=(8,6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Test')
plt.xlabel('Training Set Size')
plt.ylabel('Score')
plt.title('SVM Learning Curve')
plt.legend()
plt.show()

# Eigenfaces Visualization
fig, axes = plt.subplots(3, 5, figsize=(8, 6), subplot_kw={'xticks':[], 'yticks':[]})
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(lfw_people.images.shape[1:]), cmap='gray')
    ax.set_title(f'PC {i+1}', color='#CC00CC')
plt.suptitle('SVM First 15 Principal Components (Eigenfaces)', color='#FF5733')
plt.show()

# ---------------------------
# Prepare data for deep learning models

X_images = lfw_people.images[..., np.newaxis]
y_cat = to_categorical(y, num_classes=len(target_names))
X_train_img, X_test_img, y_train_cat, y_test_cat = train_test_split(
    X_images, y_cat, test_size=0.25, random_state=42)

print(f"Deep learning data shape: {X_train_img.shape}, {X_test_img.shape}")

cnn = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=X_train_img.shape[1:]),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(target_names), activation='softmax')
])

cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("\n=== CNN Model Summary ===")
cnn.summary()

# Train and evaluate CNN
history = cnn.fit(X_train_img, y_train_cat, epochs=10, batch_size=32,
                  validation_data=(X_test_img, y_test_cat))
cnn_eval = cnn.evaluate(X_test_img, y_test_cat, verbose=0)
print(f"CNN Test Accuracy: {cnn_eval[1]:.2f}")

# Transfer Learning with MobileNetV2
def preprocess_for_mobilenet(X):
    X_rgb = np.repeat(X, 3, axis=-1)
    X_resized = np.zeros((X.shape[0], 96, 96, 3))
    for i in range(X.shape[0]):
        img = array_to_img(X_rgb[i])
        img = img.resize((96, 96))
        X_resized[i] = img_to_array(img)
    return X_resized / 255.0

X_train_mnet = preprocess_for_mobilenet(X_train_img)
X_test_mnet = preprocess_for_mobilenet(X_test_img)

base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(96, 96, 3))
base_model.trainable = False
inputs = Input(shape=(96, 96, 3))
x = base_model(inputs, training=False)
x = GlobalAveragePooling2D()(x)
outputs = Dense(len(target_names), activation='softmax')(x)
mobilenet_model = Model(inputs, outputs)

mobilenet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("\n=== MobileNetV2 Transfer Learning Model Summary ===")
mobilenet_model.summary()

history_mnet = mobilenet_model.fit(X_train_mnet, y_train_cat, epochs=10, batch_size=32,
                                   validation_data=(X_test_mnet, y_test_cat))
mnet_eval = mobilenet_model.evaluate(X_test_mnet, y_test_cat, verbose=0)
print(f"MobileNetV2 Test Accuracy: {mnet_eval[1]:.2f}")

# --- CNN WITH CONTRASTIVE SUPERVISION (Deep Model) ---
def create_augmenter():
    return tf.keras.Sequential([
        tf.keras.layers.RandomFlip('horizontal'),
        tf.keras.layers.RandomRotation(0.1),
        tf.keras.layers.RandomZoom(0.1),
        tf.keras.layers.RandomContrast(0.15),
    ])
def projection_head(x, out_dim=128, name="projection_head"):
    x = Dense(128, activation='relu')(x)
    x = Dense(out_dim)(x)
    x = BatchNormalization()(x)
    return Lambda(lambda t: tf.math.l2_normalize(t, axis=1), name=name)(x)
def build_contrastive_cnn(input_shape, n_classes):
    inp = Input(shape=input_shape)
    x = Conv2D(32, (3,3), activation='relu', name='conv2d')(inp)
    x = MaxPooling2D(2,2)(x)
    block1_flat = Flatten(name='flatten')(x)
    proj1 = projection_head(block1_flat, name="proj1")
    x = Dense(128, activation='relu',)(block1_flat)
    x = Reshape((8,8,2))(x)
    x = Conv2D(64, (3,3), activation='relu', name='conv2d_1')(x)
    x = MaxPooling2D(2,2)(x)
    block2_flat = Flatten(name='flatten_1')(x)
    proj2 = projection_head(block2_flat, name="proj2")
    base = Dense(128, activation='relu')(block2_flat)
    feat = Dropout(0.5)(base)
    main_out = Dense(n_classes, activation='softmax', name="main_output")(feat)
    model = Model(inputs=inp, outputs=[main_out, proj1, proj2])
    return model
def nt_xent_loss(zi, zj, temperature=0.1):
    zi = tf.math.l2_normalize(zi, axis=1)
    zj = tf.math.l2_normalize(zj, axis=1)
    batch_size = tf.shape(zi)[0]
    representations = tf.concat([zi, zj], axis=0)
    similarity_matrix = tf.matmul(representations, representations, transpose_b=True)
    logits_mask = tf.ones_like(similarity_matrix) - tf.eye(2 * batch_size)
    l_pos = tf.reduce_sum(zi * zj, axis=1) / temperature
    l_neg = tf.reduce_logsumexp(similarity_matrix[:batch_size, :] * logits_mask[:batch_size, :], axis=1)
    loss = -tf.reduce_mean(l_pos - l_neg)
    return loss
def train_contrastive_supervision(model, X_train, y_train_cat, X_val, y_val, batch_size=32, epochs=8, contrastive_weight=0.1):
    augmenter = create_augmenter()
    optimizer = tf.keras.optimizers.Adam()
    ce_loss_fn = tf.keras.losses.CategoricalCrossentropy()
    ce_loss_hist, c1_loss_hist, c2_loss_hist, acc_hist = [], [], [], []
    for epoch in range(epochs):
        ce_epoch, c1_epoch, c2_epoch = [], [], []
        for i in range(0, len(X_train), batch_size):
            x_batch = X_train[i:i+batch_size]
            y_batch = y_train_cat[i:i+batch_size]
            x1_aug = augmenter(x_batch)
            x2_aug = augmenter(x_batch)
            with tf.GradientTape() as tape:
                y_pred, z1a, z2a = model(x1_aug, training=True)
                _, z1b, z2b = model(x2_aug, training=True)
                ce_loss = ce_loss_fn(y_batch, y_pred)
                c_loss1 = nt_xent_loss(z1a, z1b)
                c_loss2 = nt_xent_loss(z2a, z2b)
                total_loss = ce_loss + contrastive_weight * (c_loss1 + c_loss2)
            grads = tape.gradient(total_loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))
            ce_epoch.append(float(ce_loss)); c1_epoch.append(float(c_loss1)); c2_epoch.append(float(c_loss2))
        ce_loss_hist.append(np.mean(ce_epoch))
        c1_loss_hist.append(np.mean(c1_epoch))
        c2_loss_hist.append(np.mean(c2_epoch))
        y_pred_final, _, _ = model.predict(X_val)
        acc_epoch = np.mean(np.argmax(y_val, axis=1) == np.argmax(y_pred_final, axis=1))
        acc_hist.append(acc_epoch)
        print(f"Epoch {epoch+1}/{epochs} - CE Loss: {ce_loss_hist[-1]:.3f}, C1: {c1_loss_hist[-1]:.3f}, C2: {c2_loss_hist[-1]:.3f}, ValAcc: {acc_hist[-1]:.3f}")
    plt.figure(figsize=(8, 5))
    plt.subplot(1,2,1); plt.plot(ce_loss_hist,label='Cross-Entropy'); plt.plot(c1_loss_hist,label='Cont.Loss 1'); plt.plot(c2_loss_hist,label='Cont.Loss 2'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
    plt.title('Training Loss')
    plt.subplot(1,2,2); plt.plot(acc_hist,label='Validation Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()
    plt.title('Validation Accuracy')
    plt.tight_layout(); plt.show()
    return model

model_cds = build_contrastive_cnn(X_train_img.shape[1:], len(target_names))
model_cds = train_contrastive_supervision(model_cds, X_train_img, y_train_cat, X_test_img, y_test_cat, epochs=8)

# ROC CURVES (deep model)
final_preds, _, _ = model_cds.predict(X_test_img)
y_true = np.argmax(y_test_cat, axis=1)
plt.figure(figsize=(8,6))
for i in range(len(target_names)):
    fpr, tpr, _ = roc_curve(y_test_cat[:,i], final_preds[:,i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'CNN: {target_names[i]} (AUC={roc_auc:.2f})')
plt.plot([0,1],[0,1],'k--'); plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate'); plt.title('ROC Curves (CNN)')
plt.legend(); plt.show()

# --- ACTIVATION/FEATURE MAPS ---
def plot_activation_maps(model, layer_name, input_img):
    activation_model = Model(inputs=model.input, outputs=model.get_layer(layer_name).output)
    activations = activation_model.predict(input_img[np.newaxis, ...])
    n_filters = activations.shape[-1]
    fig, axes = plt.subplots(1, min(n_filters,8), figsize=(18,3))
    for i in range(min(n_filters,8)):
        axes[i].imshow(activations[0,:,:,i], cmap='viridis')
        axes[i].axis('off')
    plt.suptitle(f'Activation Maps: {layer_name}'); plt.show()
plot_activation_maps(model_cds, 'conv2d', X_test_img[0])

# --- GRAD-CAM ---
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    img_input = img_array[np.newaxis, ...].astype('float32')
    grad_model = Model([model.input], [model.get_layer(last_conv_layer_name).output, model.output[0]])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_input)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        output = predictions[:, pred_index]
    grads = tape.gradient(output, conv_outputs)[0]
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2)).numpy()
    conv_outputs = conv_outputs[0].numpy()
    cam = np.dot(conv_outputs, pooled_grads)
    cam = np.maximum(cam, 0)
    cam = cam / (np.max(cam)+1e-8)
    cam = cv2.resize(cam, (img_input.shape[2], img_input.shape[1]))
    cam_8u = np.uint8(255*cam)
    if cam_8u.ndim == 3: cam_8u = cam_8u[:,:,0]
    heatmap = cv2.applyColorMap(cam_8u, cv2.COLORMAP_JET)
    orig_img = img_array.squeeze()
    if orig_img.max() <= 1: orig_img = (orig_img*255).astype('uint8')
    else: orig_img = orig_img.astype('uint8')
    if orig_img.ndim == 2: orig_img = cv2.cvtColor(orig_img, cv2.COLOR_GRAY2BGR)
    superimposed_img = cv2.addWeighted(orig_img, 0.6, heatmap, 0.4, 0)
    plt.imshow(superimposed_img[...,::-1]); plt.title('Grad-CAM'); plt.axis('off'); plt.show()
make_gradcam_heatmap(X_test_img[0], model_cds, 'conv2d_1')

# --- SALIENCY MAP (with shape-fix) ---
def compute_saliency_map(model, img, class_index):
    img_tf = tf.convert_to_tensor(img[np.newaxis, ...])
    with tf.GradientTape() as tape:
        tape.watch(img_tf)
        preds = model(img_tf)[0]
        loss = preds[:, class_index]
    grads = tape.gradient(loss, img_tf)[0].numpy()
    if grads.ndim == 3:
        grads = np.max(np.abs(grads), axis=-1)
    else:
        grads = np.abs(grads)
    if grads.ndim == 1 and grads.shape[0]==img.shape[0]*img.shape[1]:
        grads = grads.reshape(img.shape[0], img.shape[1])
    plt.imshow(grads, cmap='hot'); plt.title('Saliency Map'); plt.axis('off'); plt.show()
compute_saliency_map(model_cds, X_test_img[0], np.argmax(y_test_cat[0]))

# --- t-SNE of CNN Layer Features ---
def plot_tsne_features(model, layer_name, X):
    extract = Model(model.input, model.get_layer(layer_name).output)
    feats = extract.predict(X)
    if len(feats.shape) > 2: feats = feats.reshape((feats.shape[0], -1))
    tsne = TSNE(n_components=2, random_state=42)
    X_embedded = tsne.fit_transform(feats)
    plt.scatter(X_embedded[:,0], X_embedded[:,1], c=np.argmax(y_test_cat, axis=1), cmap='rainbow', alpha=0.7)  #tab20
    plt.title(f't-SNE of {layer_name} activations'); plt.show()
plot_tsne_features(model_cds, 'flatten', X_test_img)

print("\n--- END OF ANALYSIS & VISUALIZATION ---")
```
